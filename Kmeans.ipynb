```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sb

%matplotlib inline
```

```python
df = pd.read_csv('SAUDE_PREP_1.csv')
```

```python
#sb.pairplot(df)
```

```python
# Exclui a coluna da CLASSE
X = np.array(df.drop('RESTITUIU', axis = 1))
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
df['K-classes'] = kmeans.labels_
```

```python
sb.pairplot(df,'K-classes')
```
