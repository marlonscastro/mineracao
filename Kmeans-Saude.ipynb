```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sb
from sklearn.cluster import KMeans

%matplotlib inline
```

```python
df = pd.read_csv('historico2017_2019.csv')
```

```python
#sb.pairplot(df)
```

```python
kmeans = KMeans(n_clusters=2, random_state=0, max_iter=1500)
```

```python
# Exclui a coluna da CLASSE
X = np.array(df.drop('Recuperadas', axis=1))
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
df['K-classes'] = kmeans.labels_
```

```python
sb.pairplot(df,'K-classes')
```
